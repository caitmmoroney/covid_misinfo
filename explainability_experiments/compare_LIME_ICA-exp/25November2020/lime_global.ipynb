# Make LIME do what ICA does

```python
# load packages
import pickle
import pandas as pd
```

## Load list of explanations

```python
# read in explanations list
with open('bert_explanation_list', 'rb') as f:
    explanations = pickle.load(f)
```

```python
explanations[0]
```

```python
# get unique words from all explanations
with open('tweets_vocab_list', 'rb') as f:
    vocab_list = pickle.load(f)

len(vocab_list)
```

## Get global word importances

```python
global_weights = []
for i in range(len(vocab_list)):
    targ_word = vocab_list[i]
    num_occur = 0 # number of occurrences of each word
    weight = 0 # var to store the values LIME associated with each word in each tweet
    for j in range(len(explanations)): # loop through all explanations
        explan = explanations[j]
        words = [tup[0] for tup in explan]
        vals = [tup[1] for tup in explan]
        for k in range(len(explan)): # loop through all of the entries in each explanation
            if words[k] == targ_word:
                num_occur += 1
                weight += vals[k]
    avg_weight = weight/num_occur
    #wordTup = tuple(targ_word, avg_weight)
    global_weights.append(avg_weight) # index will match index of vocab_list
```

```python
global_df = pd.DataFrame({'word': vocab_list,
                          'import': global_weights})
global_df
```

```python
unrel_words = global_df[global_df['import'] > 0]
rel_words = global_df[global_df['import'] < 0]
```
