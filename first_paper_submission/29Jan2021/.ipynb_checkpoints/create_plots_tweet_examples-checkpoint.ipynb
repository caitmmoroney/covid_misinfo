```python
import matplotlib.pyplot as plt
import matplotlib
import pickle
import numpy as np
import pandas as pd
```

# Create LIME explanation plots

```python
tweet_example_ids = [331,481,494,301,278,238,123,48]
case_example_list = ['case{}'.format(i+1) for i in range(8)]
```

```python
# load A explanations & BERT explanations
with open('bert_explanations_aslist.pkl','rb') as f:
    bert_explanations = pickle.load(f)
    
with open('a_explanations_aslist.pkl','rb') as f:
    a_explanations = pickle.load(f)
```

```python
#for i in range(1):
for i in range(len(tweet_example_ids)): # iterate over tweets in explanation list
    idx = tweet_example_ids[i]
    fname = './case_examples/tweet{}_'.format(idx)
    fname = fname + '{}_LIME.png'.format(case_example_list[i])
    
    # get BERT explanation
    bert_exp = bert_explanations[idx]
    bert_words = [tup[0] for tup in bert_exp]
    bert_vals = [tup[1] for tup in bert_exp]
    bert_df = pd.DataFrame({'BERT': bert_vals}, index = bert_words)
    
    # get ICA A matrix explanation
    ica_exp = a_explanations[idx]
    ica_words = [tup[0] for tup in ica_exp]
    ica_vals = [tup[1] for tup in ica_exp]
    ica_df = pd.DataFrame({'ICA': ica_vals}, index = ica_words)
    
    # combine into dataframe
    full_df = bert_df.merge(ica_df, left_index = True, right_index = True)
    
    # get x tick labels
    xlabels = list(full_df.index)
    
    # adjust font size settings
    font = {'family' : 'sans-serif',
            'size': 14,
            'weight': 400}

    matplotlib.rc('font', **font)
    
    # plot values
    ax = full_df.plot.bar()#rot = 50)
    plt.grid(True, linewidth = 0.3)
    ax.set_xticklabels(xlabels, rotation = 50, ha = 'right')
    plt.savefig(fname, dpi = 300, bbox_inches = 'tight')
```

# Extract top 15 words from top 3 topics

```python
tweet_embeddings = np.load('tweet_embed_A.npy')
```

```python
topic_list = ['Topic{}'.format(i+1) for i in range(tweet_embeddings.shape[1])]
```

```python
topic_array = np.array(topic_list)
```

```python
# sort all tweets at once (sort by magnitude); get top 3 topics
tweet_topics_mag = np.flip(np.argsort(abs(tweet_embeddings), 1), 1)[:,:3]
tweet_topic_names_mag = topic_array[tweet_topics_mag]
```

```python
# read in topic-word table
topic_words = pd.read_csv('topics_by_magnitude.csv', index_col = 0)
```

```python
for i in range(len(tweet_example_ids)): # iterate over tweets in explanation list
    idx = tweet_example_ids[i]
    fname = './case_examples/tweet{}_'.format(idx)
    fname = fname + '{}_top3topics.csv'.format(case_example_list[i])
    
    # get top 10 words from top 3 topics
    df = topic_words[tweet_topic_names_mag[idx].tolist()][:10]
    df.to_csv(fname)
```
